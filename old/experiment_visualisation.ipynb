```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_graph(df, name):
    plt.figure(figsize=(12.2, 4.5))
    x = np.arange(df.shape[0])
    for i in df.columns:
        plt.plot(x, df[i], color ='black', linestyle ='dashed', linewidth=1)
    plt.title(name)
    plt.xlabel('Iterations')
    plt.ylabel('Profit values')
    plt.xlim(0,500)
    name_of_graph = name.lower()+'_experiment_graph'
    # plt.savefig('name_of_graph', bbox_inches='tight')
    plt.show()
```

```python
def get_stats(df):
    _max = df.iloc[-1].max()
    _min = df.iloc[-1].min()
    _mean = df.iloc[-1].mean()
    _median = np.median(df.iloc[-1])
    _stddev = round(df.iloc[-1].std(),2)
    return [_max, _min, _mean, _median, _stddev]

def get_iter_stats(df):
    _max = df.idxmax().max()
    _min = df.idxmax().min()
    _mean = round(df.idxmax().mean(),2)
    _median = round(np.median(df.idxmax()),2)
    _stddev = round(df.idxmax().std(),2)
    return [_max, _min, _mean, _median, _stddev]
```

```python
cpso_rd = pd.read_excel('experiment_result_cpso_reduced_supply3.xlsx', index_col=0, nrows=502)
```

```python
plot_graph(cpso_rd[:-1], 'CPSO reduced supply')
```

```python
sib_rd = pd.read_excel('experiment_result_sib_reduced_supply_mix10_jump10.xlsx', index_col=0)
```

```python
plot_graph(sib_rd[:-1], 'SIB reduced supply')
```

```python
def first_last_iter(col)-> bool:
    first_row = 0
    last_row = -1
    return col[first_row] == col[last_row]
```

```python
sibv_rd = pd.read_excel('experiment_result_sibv_reduced_supply_mix5_jump20.xlsx', index_col=0, nrows=502)
```

```python
plot_graph(sibv_rd[:-1], 'SIBv reduced supply')
```

```python
summary_table = pd.DataFrame(columns=['CPSO', 'SIB', 'SIBv'], index=['Max', 'Min', 'Mean', 'Median', 'StdDev'])
```

```python
summary_table['CPSO'] = get_stats(cpso_rd[:-1])
summary_table['SIB'] = get_stats(sib_rd[:-1])
summary_table['SIBv'] = get_stats(sibv_rd[:-1])
```
