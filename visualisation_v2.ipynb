```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sib_ex = pd.read_excel('experiment_result_sib_ex.xlsx', index_col=0, nrows=500, converters={1:float, 2:float, 3:float, 4:float, 5:float} )
```

```python
def plot_graph(df, name):
    plt.figure(figsize=(12.2, 4.5))
    x = np.arange(df.shape[0])
    for i in df.columns:
        plt.plot(x, df[i], color ='black', linestyle ='dashed', linewidth=1)
    plt.title(name)
    plt.xlabel('Iterations')
    plt.ylabel('Profit values')
    # plt.xlim(0,500)
    name_of_graph = name.lower()+'_ex_experiment_graph'
    # plt.savefig('name_of_graph', bbox_inches='tight')
    plt.show()
```

```python
def get_stats(df):
    _max = df.iloc[-1].max()
    _min = df.iloc[-1].min()
    _mean = df.iloc[-1].mean()
    _median = np.median(df.iloc[-1])
    _stddev = round(df.iloc[-1].std(),2)
    return [_max, _min, _mean, _median, _stddev]
```

```python
def get_iter_stats(df):
    _max = df.idxmax().max()
    _min = df.idxmax().min()
    _mean = round(df.idxmax().mean(),2)
    _median = round(np.median(df.idxmax()),2)
    _stddev = round(df.idxmax().std(),2)
    return [_max, _min, _mean, _median, _stddev]
```

```python
plot_graph(sib_ex, 'SIB')
```

```python
sibv_ex = pd.read_excel('experiment_result_sibv_ex.xlsx', index_col=0, nrows=500, converters={1:float, 2:float, 3:float, 4:float, 5:float} )
```

```python
plot_graph(sibv_ex, 'SIBv')
```

```python
cpso_ex = pd.read_excel('experiment_result_cpso_ex.xlsx', index_col=0, nrows=500, converters={1:float, 2:float, 3:float, 4:float, 5:float} )
```

```python
plot_graph(cpso_ex, 'CPSO+')
```

```python
summ_ex = pd.DataFrame(columns=['CPSO', 'SIB', 'SIBv'], index=['Max', 'Min', 'Mean', 'Median', 'StdDev'])
```

```python
summ_ex.loc['Max']['CPSO'] = cpso_ex.iloc[-1][:].max()
summ_ex.loc['Max']['SIB'] = sib_ex.iloc[-1][:].max()
summ_ex.loc['Max']['SIBv'] = sibv_ex.iloc[-1][:].max()

summ_ex.loc['Min']['CPSO'] = cpso_ex.iloc[-1][:].min()
summ_ex.loc['Min']['SIB'] = sib_ex.iloc[-1][:].min()
summ_ex.loc['Min']['SIBv'] = sibv_ex.iloc[-1][:].min()

summ_ex.loc['Mean']['CPSO'] = cpso_ex.iloc[-1][:].mean()
summ_ex.loc['Mean']['SIB'] = sib_ex.iloc[-1][:].mean()
summ_ex.loc['Mean']['SIBv'] = sibv_ex.iloc[-1][:].mean()

summ_ex.loc['Median']['CPSO'] = cpso_ex.iloc[-1][:].median()
summ_ex.loc['Median']['SIB'] = sib_ex.iloc[-1][:].median()
summ_ex.loc['Median']['SIBv'] = sibv_ex.iloc[-1][:].median()

summ_ex.loc['StdDev']['CPSO'] = cpso_ex.iloc[-1][:].std()
summ_ex.loc['StdDev']['SIB'] = sib_ex.iloc[-1][:].std()
summ_ex.loc['StdDev']['SIBv'] = sibv_ex.iloc[-1][:].std()
```

```python
summ_ex_iters = pd.DataFrame(columns=['CPSO', 'SIB', 'SIBv'], index=['Max', 'Min', 'Mean', 'Median', 'StdDev'])
```

```python
summ_ex_iters['CPSO']= get_iter_stats(cpso_ex)
summ_ex_iters['SIB']= get_iter_stats(sib_ex)
summ_ex_iters['SIBv']= get_iter_stats(sibv_ex)
```
